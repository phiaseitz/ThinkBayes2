```python
import thinkbayes2
import thinkplot
```

```python
pmf = thinkbayes2.NormalPdf(0,1).MakePmf()
```

```python
thinkplot.Pdf(pmf)
```

```python
pmfX = thinkbayes2.NormalPdf(163,7.3).MakePmf()
pmfY = thinkbayes2.NormalPdf(177,7.7).MakePmf()
```

```python
thinkplot.Pdf(pmfX)
thinkplot.Pdf(pmfY)
```

```python
n = 1000
xs = pmfX.MakeCdf().Sample(n)
ys = pmfY.MakeCdf().Sample(n)
zs = xs + ys
pmfZ = thinkbayes2.Pmf(zs)
thinkplot.Cdf(pmfZ.MakeCdf())
```

```python
pmfZ = pmfX + pmfY
thinkplot.Pdf(pmfZ)
```

```python
def PmfProduct(pmf1, pmf2):
    """Product of two PMFs"""
    pmf = thinkbayes2.Pmf()
    for v1,p1 in pmf1.Items():
        for v2,p2 in pmf2.Items():
            pmf.Incr(v1*v2, p1*p2)
    return pmf
```

```python
pmfz = PmfProduct(pmfX, pmfY)
thinkplot.Cdf(pmfz.MakeCdf())
```

```python
n = 1000
xs = pmfX.MakeCdf().Sample(n)
ys = pmfY.MakeCdf().Sample(n)
zs = xs * ys
pmfZ = thinkbayes2.Pmf(zs)
thinkplot.Cdf(pmfZ.MakeCdf())
```

```python
pmfZ = pmfX + pmfY
def PmfProbSumGreater(pmf1, pmf2, pmf3):
    pmf4 = pmf1 + pmf2
    p = pmf4.ProbGreater(pmf3)
    thinkplot.Cdfs([pmf3.MakeCdf(), pmf4.MakeCdf()])
    return p
PmfProbSumGreater(pmfX, pmfY, pmfZ)
```
